```python
def autenticar():
    load_dotenv('env/autenticar.env')
    
    usuario = os.getenv('USUARIO')
    senha = os.getenv('SENHA')
    
    url = 'https://integra.ons.org.br/api/autenticar'
    data = '{"usuario" : "'+usuario+'", "senha": "'+senha+'"}'
    headers = {'accept': 'application/json', 'Content-Type': 'application/json'}
    
    return requests.post('https://integra.ons.org.br/api/autenticar', headers=headers, data=data).json()['access_token']
```

```python
access_token = autenticar()
```

```python
access_token
```

### Requisitando dados da API

```python
headers = {
    'accept': 'application/json',
    'Authorization': f'Bearer {access_token}'}

params = (
    ('cod_areacarga', 'SP'),
    ('dat_referencia_ini', '2016-01-01'),
    ('dat_referencia_fim', '2016-04-31'),
)

response = requests.get('https://integra.ons.org.br/api/cargaglobal/cargav', headers=headers, params=params)
response
```

```python
response.json()
```

```python
def requisitar(access_token, ini, fim):
    headers = {
    'accept': 'application/json',
    'Authorization': f'Bearer {access_token}'}

    params = (
        ('cod_areacarga', 'SP'),
        ('dat_referencia_ini', f'{ini}'),
        ('dat_referencia_fim', f'{fim}'),
    )
    
    return requests.get('https://integra.ons.org.br/api/cargaglobal/cargav', headers=headers, params=params).json()
```

```python
requisitar(access_token, '2021-01-01', '2021-01-31')
```

### Criar colunas de data e hora corrigindo o horário para o fuso UTC-3

```python
def dia(x):
    y = re.sub('Z', '', x)
    y = (datetime.datetime.fromisoformat(y)+datetime.timedelta(hours=-3)).strftime("%d-%m-%Y")
    return y
```

```python
dia('2021-01-01T03:30:00.000Z')
```

```python
def hora(x):
    y = re.sub('Z', '', x)
    y = (datetime.datetime.fromisoformat(y)+datetime.timedelta(hours=-3)).strftime("%H:%M")
    return y
```

```python
hora('2021-01-01T03:30:00.000Z')
```

```python
df['dia'] = df.din_referenciautc.map(dia)
df['hora_UTC-3'] = df.din_referenciautc.map(hora)
df
```

```python
def timestamp(x):
    y = re.sub('Z', '', x)
    y = (datetime.datetime.fromisoformat(y)+datetime.timedelta(hours=-3)).strftime("%d-%m-%Y %H:%M")
    return y
```

### Criar coluna com o dia da semana

```python
def dia_da_semana(x):
    y = re.sub('Z', '', x)
    y = (datetime.datetime.fromisoformat(y)+datetime.timedelta(hours=-3)).weekday()
    weekday = {0: '2. Segunda-Feira', 1: '3. Terça-Feira', 2: '4. Quarta-Feira', 3 : '5. Quinta-Feira', 4 : '6. Sexta-Feira', 5 : '7. Sábado', 6 : '1. Domingo'}
    return weekday[y]
```

```python
df['dia_semana'] = df.din_referenciautc.map(dia_da_semana)
df
```

```python
df['Carga_MW'] = df.val_cargaglobal
df
```

## Juntando tudo
### Puxar dados desde 2016

```python
ini_dates = ['01-01', '05-01', '09-01']
end_dates = ['04-31', '08-31', '12-31']
anos = ['2016', '2017', '2018', '2019', '2020']
```

```python
access_token = autenticar()
```

```python
df_list = []

for ano in anos:
    for i in range(0, len(ini_dates)):
        json = requisitar(access_token, ano+'-'+ini_dates[i], ano+'-'+end_dates[i])
        df_list.append(pd.DataFrame(json))
```

```python
df_list.append(pd.DataFrame(requisitar(access_token, '2021-01-01', '2021-04-31')))
df_list.append(pd.DataFrame(requisitar(access_token, '2021-05-01', '2021-08-31')))
df_list.append(pd.DataFrame(requisitar(access_token, '2021-09-01', '2021-10-11')))
```

```python
len(df_list)
```

```python
df_carga = pd.concat(df_list, ignore_index=True)
```

```python
df_carga
```

```python
df_carga['timestamp_UTC-3'] = df_carga.din_referenciautc.map(timestamp)
df_carga['Carga_MW'] = df_carga.val_cargaglobal
df_carga.drop(columns=['dat_referencia', 'din_referenciautc', 'val_cargaglobal', 'val_cargaglobalcons', 'val_consistencia', 'val_cargasup', 'val_cargansup'], inplace=True)
```
